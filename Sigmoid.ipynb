##**Sigmoid function**

```python
import numpy as np
import matplotlib.pyplot as plt

e=np.e

def sigmoid(x):
  s=1/(1+e**-x) #can also be written as s=1/(1+np.exp(-x))
  return s

def step_function(x):
  y=x>0
  return y.astype(np.int)

x1=np.linspace(-5,5)
y_sig=sigmoid(x1)

x2=np.arange(-5.0,5.0,0.0001)
y_step=step_function(x2)

plt.plot(x1,y_sig,label ="sigmoid")
plt.plot(x2,y_step, linestyle = "dashed",label ="step")
plt.legend()

plt.xlabel("x",size=14)
plt.ylabel("y",size=14)
plt.grid()
plt.show()
```

$$y = \frac{1}{1+e^{-x}}$$ 
$$y' = y(1-y)$$ 


```python
import numpy as np
import matplotlib.pyplot as plt
dx = 0.1
e=np.e

def sigmoid(x):
  s=1/(1+np.exp(-x))
  return s

def df_sigmoid(x):
  d=sigmoid(x)*(1-sigmoid(x))
  return d

y_sig=sigmoid(x)
y_d=(sigmoid(x+dx)-sigmoid(x))/dx
y_df=df_sigmoid(x)

plt.plot(x,y_sig,label="sigmoid")
plt.plot(x,y_d,label="d")
plt.plot(x,y_df,label="df")
plt.legend()

plt.xlabel("x",size=14)
plt.ylabel("y",size=14)
plt.grid()
plt.show()
```
