##Drawing equations
$$y = 2x+1$$

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-5,5)
y =2*x+1

plt.plot(x,y)
plt.show()
```

$$y=x^2$$
$$y=1/x$$
$$y=0.5x^3-6x$$

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-5,5)
y1 =x**2
y2=1/x
y3 = 0.5*(x**3)-6*x

plt.xlabel("x value", size =10)
plt.ylabel("y value", size = 10)

plt.title ("Functions")
plt.grid()

plt.plot(x,y1, label = "y1")
plt.plot(x,y2, label = "y2", linestyle = "dashed")
plt.plot(x,y3, label = "y3", linestyle = "dashdot")
plt.legend()
plt.show()
```

$$y=2^x$$
$$y=3^x$$

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-2,2)
y4 =2**x
y5=3**x

plt.xlabel("x", size =10)
plt.ylabel("y", size = 10)

plt.grid()

plt.plot(x,y4, label = "y4")
plt.plot(x,y5, label = "y5", linestyle = "dashed")

plt.legend()
plt.show()
```

$$y=e^x$$

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-2,2)
e = np.e

y_e = e**x
y4 =2**x
y5=3**x

plt.xlabel("x", size =10)
plt.ylabel("y", size = 10)

plt.grid()

plt.plot(x,y_e, label = "e^x")
plt.plot(x,y4, label = "2^x", linestyle = "dashdot")
plt.plot(x,y5, label = "3^x", linestyle = "dashed")

plt.legend()
plt.show()
```
